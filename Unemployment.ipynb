```python
import pandas as pd
```

```python
unemploy_raw = pd.read_csv("Resources/Unemployment Rate.csv")
unemploy_raw.head()
```

```python
#Change Month column's data type into integer

unemploy_raw["Month"] = unemploy_raw["Month"].map({"January": 1, "February": 2, "March": 3, "April": 4, "May": 5,
                                "June": 6, "July": 7, "August": 8, "September": 9, "October": 10,
                                "November": 11, "December": 12})
unemploy_raw.head()
```

```python
#Slice for 2016 data only
unemploy_2016 = unemploy_raw[unemploy_raw['Year']==2016]
unemploy_2016.head()
```

```python
#Drop unused column
unemployment = unemploy_2016.drop(columns = ['County'])
unemployment.reset_index()
```

## Load into PostgreSQL database

```python
from sqlalchemy import create_engine
etl_connection_string = "postgres:postgres@localhost:5432/etlproject_db"
engine = create_engine(f'postgresql://{etl_connection_string}')
```

```python
engine.table_names()
```

```python
unemployment.to_sql(name = 'unemployment', con = engine, if_exists='replace', index=False)
```

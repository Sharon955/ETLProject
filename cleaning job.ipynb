```python
import pandas as pd
```

```python
raw = pd.read_csv("./Resources/US_jobs.csv")
raw.head()
```

```python
new = raw.drop(['job_board','job_description','page_url','phone_number','salary','job_type','sector'],axis=1)
new[['month','date','year']] = new.date_added.str.split("/",expand=True,)
new = new.drop(['date_added','date','country_code'],axis=1)
new.head()
```

```python
new['state']=new.location.str[-2:]
new.head()
```

```python
from sqlalchemy import create_engine
etl_connection_string = "postgres:postgres@localhost:5432/etlproject_db"
engine = create_engine(f'postgresql://{etl_connection_string}')
```

```python
engine.table_names()
```

```python
new.to_sql(name='joblisting', con=engine, if_exists='append', index=False)
```
